```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
x=data.iloc[:,[3,4]]
print(x.head())
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x, method = "ward"))
plt.title("dendrogram")
plt.xlabel("customers")
plt.ylabel("euclidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5, linkage="ward")
# affinity is not required as euclidean distance is default for linkage
y_hc=hc.fit_predict(x)
print(y_hc)
```

```python
plt.scatter(x.iloc[y_hc==0 ,0], x.iloc[y_hc==0 , 1], s=100, color="red", label="cluster1")
plt.scatter(x.iloc[y_hc==1 ,0], x.iloc[y_hc==1 , 1], s=100, color="blue", label="cluster2")
plt.scatter(x.iloc[y_hc==2 ,0], x.iloc[y_hc==2 , 1], s=100, color="yellow", label="cluster3")
plt.scatter(x.iloc[y_hc==3 ,0], x.iloc[y_hc==3 , 1], s=100, color="green", label="cluster4")
plt.scatter(x.iloc[y_hc==4 ,0], x.iloc[y_hc==4 , 1], s=100, color="purple", label="cluster5")
plt.title("cluster of customers")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```

**We can also set 3 as clusters as this will also be the optimal numbers of clusters according the dendrogram**
