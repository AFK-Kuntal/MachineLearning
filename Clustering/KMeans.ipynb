```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
x=data.iloc[: , [3,4]]
print(x)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  model=KMeans(n_clusters = i , init = "k-means++" , random_state = 42 )
  model.fit(x)
  wcss.append(model.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.title("WCSS")
plt.show()
```

```python
 model=KMeans(n_clusters = 5 , init = "k-means++" , random_state = 42 )
 y_kmeans=model.fit_predict(x)
 y_kmeans
```

```python
plt.scatter(x.iloc[y_kmeans == 0 , 0] , x.iloc[y_kmeans == 0 , 1] , s = 100 , c = "blue" , label = "cluster1")
plt.scatter(x.iloc[y_kmeans == 1 , 0] , x.iloc[y_kmeans == 1 , 1] , s = 100 , c = "red" , label = "cluster2")
plt.scatter(x.iloc[y_kmeans == 2 , 0] , x.iloc[y_kmeans == 2 , 1] , s = 100 , c = "yellow" , label = "cluster3")
plt.scatter(x.iloc[y_kmeans == 3 , 0] , x.iloc[y_kmeans == 3 , 1] , s = 100 , c = "green" , label = "cluster4")
plt.scatter(x.iloc[y_kmeans == 4 , 0] , x.iloc[y_kmeans == 4 , 1] , s = 100 , c = "magenta" , label = "cluster5")
plt.scatter(model.cluster_centers_[:,0] , model.cluster_centers_[:,1] , s = 300 , c = "black" , label = "centeroid")
plt.title("KMeans clusters")
plt.xlabel("Annual Income")
plt.xlabel("spending score")
plt.legend()
plt.show()
```
